```python
import pandas as pd
import xml.etree.ElementTree as ET
import os
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.lm import Vocabulary
import matplotlib.pyplot as plt
```

```python
folder = "eduardo/opus libros es/Books/raw/es"
```

```python
xml_files = os.listdir(folder)
xml_files
```

```python
book = ET.parse(os.path.join(folder, xml_files[0]))
book
```

```python
for sentence in book.getroot().iter("s"):
    print(sentence.text)
```

```python
roots = pd.Series([ET.parse(os.path.join(folder, f)).getroot() for f in xml_files])
roots
```

```python
sentences = pd.DataFrame(pd.Series(xml_files), columns = ["file"])\
.assign(root = lambda df: df["file"].map(lambda x: ET.parse(os.path.join(folder, x)).getroot()))\
.assign(sentences = lambda df: df["root"].map(lambda x: list(enumerate([y.text for y in x.iter("s")]))))\
.drop("root", axis = 1)\
.set_index("file")\
["sentences"]\
.explode()\
.reset_index()\
.assign(order = lambda df: df["sentences"].map(lambda x: x[0]),
        text = lambda df: df["sentences"].map(lambda x: x[1]))\
.drop("sentences", axis = 1)

sentences
```

# exploratory data analysis

## character counts

```python
char_counts = pd.Series(Vocabulary("".join(sentences["text"].to_list())).counts).sort_values(ascending = False)
char_counts.head(10)
```

```python
"{:,}".format(char_counts.sum())
```

```python
char_counts.describe().map(lambda x: "{:,.1f}".format(x))
```

```python
plt.figure(figsize = (15, 5))
char_counts.plot.hist(bins = 200)
```

```python
plt.figure(figsize = (15, 5))
char_counts.plot.hist(bins = 200, log = True)
```

```python
plt.figure(figsize = (18, 30))
char_counts.plot.barh()
```
